```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
#Store the data 
df = pd.read_csv('udemy_tech.csv')
# show the first 3 rows of data 
df.head(3)
```

```python
df.shape
```

```python
# Creat a list of important columns for the recommendation engine
columns= ['Title','Summary']
```

```python
# show the data 
df[columns].head(3)
```

```python
# check for missing values in the important columns
df[columns].isnull().values.any()

# data cleaning is needed
```

```python
# Create a function to combine the values of the important columns into a single string
def get_important_features(data):
  important_features = []
  for i in range(0, data.shape[0]):
    important_features.append(str(data['Title'][i])+' '+str(data['Summary'][i]))
  return important_features
```

```python
# Create a column to hold the combined strings
df['important_features'] = get_important_features(df)

#show the data 
df.head(3)
```

```python
# creating stop word
stop_words = stopwords.words('english') + stopwords.words('french')
print(stop_words)
```

```python
# convert the text to a matrix of token counts 
df['important_features'] = df['important_features'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
cm = CountVectorizer().fit_transform(df['important_features'])

# GEt the cosine similarity matrix from the count matrix
cs = cosine_similarity(cm)
#Print the cosine similarity matrix
print(cs)
```

```python
# Get the shape of the cosine similarity matrix 
cs.shape
```

```python
# Get the title of the course that the user likes 
title = 'iOS 13 & Swift 5 - The Complete iOS App Development Bootcamp'

# Find the course id
course_id = df[df.Title == title]['Course_id'].values[0]
```

```python
# Create a list of enumeration for the similarity score [(course_id, similiraty score ),(..)]
scores = list(enumerate(cs[course_id]))
```

```python
#sort the list 
sorted_scores = sorted(scores, key = lambda x:x[1], reverse  = True)
sorted_scores = sorted_scores[1:]
```

```python
#Print the sorted scores
print(sorted_scores)
```

```python
# Create a loop to print the first 7 similar movies
j = 0
print('The 7 most recommended courses', title, 'are:\n')
print('ID  Course Title' )
for item in sorted_scores:
  course_title = df[df.Course_id == item[0]]['Title'].values[0]
  course_id = df[df.Course_id == item[0]]['Course_id'].values[0]
  print(course_id, course_title)
  j=j+1
  if j>100:
    break
```
